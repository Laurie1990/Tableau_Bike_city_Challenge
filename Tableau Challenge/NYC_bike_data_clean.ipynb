```python
import pandas as pd
```

```python
import numpy as np
```

```python
bike_path='202010-citibike-tripdata'
```

```python
bike_data=pd.read_csv(bike_path)
```

```python
bike_data.head()
```

```python
bike_data.shape
```

```python
#Hmmm, let's drop all data from before 2020
bike_data.dtypes
```

```python
#Split start time into date and time columns, and then filter on date
new = bike_data["starttime"].str.split(" ", n = 1, expand = True) 
```

```python
new.head()
```

```python
bike_data["date"]=new[0]
```

```python
bike_data["time"]=new[1]
```

```python
bike_data = bike_data.astype({"date": "object", "time": "datetime64"})
```

```python
bike_data.head()
```

```python
bike_data.dtypes
```

```python
bike_data.drop(bike_data[bike_data["date"]<"2020-09-15"].index, inplace=True)
```

```python
bike_data.shape
```

```python
bike_data.loc[:, "birth year"] = bike_data["birth year"].apply(lambda x: 2020-x)
```

```python
bike_data.head()
```

```python
bike_data.rename(columns = {'birth year':'age'}, inplace = True)
```

```python
bike_data.head()
```

```python
bike_data['trip_id']=bike_data.index
bike_data = bike_data.reset_index(drop=True)
```

```python
bike_data.head()
```

```python
end_point = bike_data[['stoptime', 'end station latitude', 'end station longitude', 'trip_id']].copy()
```

```python
end_point.head()
```

```python
bike_data.to_csv('big_nyc_cleaned_data.csv', index = False)
```

```python
end_point.to_csv('big_nyc_cleaned_endpoint.csv', index = False)
```
